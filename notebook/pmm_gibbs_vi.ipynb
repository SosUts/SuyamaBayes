```julia
using BenchmarkTools
using Distributions
using PyPlot

using SuyamaBayes
```

Gibbs sampling

```julia
a = 1.0
b = 0.01
N = 1000
true_位 = [100., 50., 15.]
true_p = [0.5, 0.2, 0.3]
true_model, toy_data = generate_toy_data(N)
prior, model = init_model(a = a, b = b)

pmm = PoissionMixtureModel()
pmm.model = model
```

```julia
gibbs!(pmm, prior, toy_data, maxiter=4000);
```

```julia
fig, ax = plt.subplots(3, 1, figsize=(5, 7))
K = length(probs(pmm.model))
ax[1].set(title="Posterior distribution of 位")
for k = 1:K
    ax[1].hist(pmm.位s[2000:end, k], density=1, bins=20)
end
ax[1].vlines(true_位, 0, ax[1].get_ylim()[2], color="k", linestyle="--")

ax[2].set(title="Posterior distribution of p", xlim=(0, 1))
for k = 1:K
    ax[2].hist(pmm.ps[2000:end, k], density=1, bins=20)
end
ax[2].vlines(true_p, 0, ax[2].get_ylim()[2], color="k", linestyle="--")

ax[3].set(title="Log likelihood", xlabel="Iteration")
ax[3].plot(1:4000, pmm.loglikelihoods)

plt.tight_layout()
```

Variational inference

```julia
a = 1.
b = 0.01
N = 100
true_位 = [100., 50., 15.]
true_p = [0.5, 0.2, 0.3]
true_model, toy_data = generate_toy_data(N)
prior, model = init_model(a = a, b = b)

vi!(prior, toy_data, maxiter=100)
```

```julia
K = length(prior.a)
fig, ax = plt.subplots(2, 1, figsize=(5, 7))
ax[1].set(title="Posterior distribution of 位")
ax[2].set(title="Posterior distribution of p", xlim=(0, 1))

post_位 = [rand(Gamma(prior.a[k], 1/prior.b[k]), 1000) for k = 1:3]
post_p = rand(Dirichlet(prior.伪), 1000)

for k = 1:K
    ax[1].hist(post_位[k], density=1)
    ax[2].hist(post_p[k, :], density=1)
end
ax[1].vlines(true_位, 0, ax[1].get_ylim()[2], color="k", linestyle="--")
ax[2].vlines(true_p, 0, ax[2].get_ylim()[2], color="k", linestyle="--")
```
